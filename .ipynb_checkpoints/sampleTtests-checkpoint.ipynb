```python
# Dependencies and Setup
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
incomedata = pd.read_csv("incomeschooldata.csv")
```

```python
incomedata.head(10)
```

```python
GAdata = incomedata.loc[(incomedata['State_y'] == 'GA')]
GAincome = GAdata['2018']

TNdata = incomedata.loc[(incomedata['State_y'] == 'TN')]
TNincome = TNdata['2018']

ALdata = incomedata.loc[(incomedata['State_y'] == 'AL')]
ALincome = ALdata['2018']

MSdata = incomedata.loc[(incomedata['State_y'] == 'MS')]
MSincome = MSdata['2018']

FLdata = incomedata.loc[(incomedata['State_y'] == 'FL')]
FLincome = FLdata['2018']

NCdata = incomedata.loc[(incomedata['State_y'] == 'NC')]
NCincome = NCdata['2018']

SCdata = incomedata.loc[(incomedata['State_y'] == 'SC')]
SCincome = SCdata['2018']

LAdata = incomedata.loc[(incomedata['State_y'] == 'LA')]
LAincome = LAdata['2018']
```

```python
incomedata['State_y'].value_counts()
```

```python
population = incomedata['2018']
sample = GAincome
    
# Scatter Plot of Data
plt.subplot(2, 1, 1)
plt.scatter(range(len(population)), population, label="population")
plt.scatter(range(len(sample)), sample, label="sample")
plt.title('GA Incomes vs. SouthEast Regional Incomes')
plt.legend()

# Histogram Plot of Data
plt.subplot(2, 1, 2)
plt.hist(population, 20, density=True, alpha=0.7, label="population")
plt.hist(sample, 20, density=True, alpha=0.7, label="sample")
plt.axvline(population.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(sample.mean(), color='k', linestyle='dashed', linewidth=1)
plt.legend()  

stats.ttest_1samp(sample, population.mean())
```
