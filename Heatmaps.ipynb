```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import requests
from pprint import pprint
from pandas import json_normalize
import json
import gmaps
from api_key import gkey
from api_key import okey
```

```python
newindex= pd.read_csv(("incomeschooldata.csv"))
newindexdf=pd.DataFrame(newindex)
```

```python
cities=newindexdf['City']
cities1=cities.reset_index()
cities2=cities1.City
```

```python
#option1
url = "http://api.openweathermap.org/data/2.5/weather?"
lat=[]
lng=[]
units="metric"
name1=[]

#Loodk for city in cities data frame in openweathermap and create new dataframe
# Build query URL

for city in cities2:
    query_url = url + "appid=" + okey + "&q=" + city +"&units="+ units
    response = requests.get(query_url).json()
    try:
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        name1.append(response['name'])
        
    except KeyError:
        print("city not found")
        
```

```python
city_data1=pd.DataFrame({"City": name1,"lat": lat,"long": lng})
city_data1
newmergedata=pd.merge(newindexdf,city_data1, on="City")
# county_df = combineddata.sort_values(
#     ["County"], ascending=True)
# county_df.head(20)

top20= newmergedata.sort_values(['RankStatewidePercentage'], ascending=False).head(20)
top20
```

```python
location2 = newmergedata[["lat", "long"]]
Income = newmergedata["2018"]

fig1 = gmaps.figure()
Income_layer = gmaps.heatmap_layer(location2, weights= Income) 
                                   #dissipating=False, max_intensity=100000, point_radius=1)
#Add layer
fig1.add_layer(Income_layer)
# Display figure
fig1
```
